```python
from bayes_nanospace2025.bayesian_search import BayesianSearch, BayesianSearchWidget, LowerConfidenceBound, LowerConfidenceBoundExponent, ExpectedImprovement, ProbabilityOfImprovement
from agox.models.descriptors import Fingerprint
from bayes_nanospace2025.descriptors import ReactionCoordinateDescriptor, CartesianCoordinateDescriptor, NearestDistanceDescriptor
from bayes_nanospace2025 import get_initial_training_data
from bayes_nanospace2025.model import make_model
import matplotlib.pyplot as plt

from agox.utils.ray import ray_startup

ray_startup(1)
```

```python
training_data, training_energies = get_initial_training_data()

descriptors = []

descriptors.append(ReactionCoordinateDescriptor(factor=100))
descriptors.append(CartesianCoordinateDescriptor())
descriptors.append(NearestDistanceDescriptor())
# descriptors.append(Fingerprint(training_data[0]))

acquisition_functions = []
acquisition_functions.append(LowerConfidenceBound(kappa=2))
acquisition_functions.append(LowerConfidenceBoundExponent(kappa=2, exponent=1))
acquisition_functions.append(ProbabilityOfImprovement(xi=0.01))
acquisition_functions.append(ExpectedImprovement(xi=0.01))
acquisition_function = acquisition_functions[0]

descriptor = descriptors[0]
model = make_model(descriptor)
bayes = BayesianSearch(model=model, acquisition=acquisition_function)
```

```python
BayesianSearchWidget(bayes, descriptors=descriptors, acquisition_functions=acquisition_functions)
```
