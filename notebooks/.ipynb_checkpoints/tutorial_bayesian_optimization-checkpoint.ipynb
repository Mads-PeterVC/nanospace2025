### Gaussian Process Regression

Clearly we need a better strategy. Our approach will involve a Gaussian Process Regression (GPR) model and an active learning
strategy in which the GPR itself chooses the next point at which to do an actual calculation. GPR predictions are computed
using the following expression

$$E(\mathbf{x}) = \sum_i \alpha_i k(\mathbf{x}, \mathbf{x}_i)$$

Where $\mathbf{x}$ is a vector of coordinates (or features) and $k(\mathbf{x}, \mathbf{x}_i)$ is a kernel-function evaluated 
between the input vector $\mathbf{x}$ and the vector of training data point $\mathbf{x}_i$, lastly $\alpha_i$ is a coefficient that is found during the training procedure.

We will now implement our own simple Gaussian process regression code. For more information on Gaussian processes see [Gaussian Process for Machine Learning](http://gaussianprocess.org/gpml/chapters/RW.pdf)

First, we will implement a radial basis kernel aka. Gaussian kernel. 
The kernel function should accept one or two feature matrices as well as a length scale and then return the kernel matrix. 

The expression for the kernel element $k(X,X')_{ij}$ is

$$k(X,X')_{ij} = \exp\left(\frac{-d(X_i, X'_j)}{2\lambda^2}\right),$$

with $d(.,.)$ the squared Euclidean distances between the two features and $\lambda$ the length-scale of the kernel. 

We will multiply with a constant of $100$ in order to be able to interpret the kernel as similarities between features as a percentage.
Below you will find scaffold code for the kernel function with a missing line, which you should fill out. 
Kernels can be thought of as a similarity functions that output high values for similar inputs and low values for dissimilar input. 

```python
from scipy.spatial.distance import cdist
def kernel(X, Y=None, ls=0.5):
  if Y is None:
    Y = X

  d = cdist(X, Y, metric='sqeuclidean')
  K = 100*np.exp(-0.5 * d / ls**2) #K = 100 * #YOUR CODE HERE
  return K
```

In order to test our kernel we will need to choose a descriptor, that is some way of generating a vector describing the atomic structures we are working with. 

At first we will try a rather simple one - namely one that uses the pathway coordinate. Let us generate training data (X,Y) for a couple of structures.

```python
from bayes_nanospace2025.descriptors import ReactionCoordinateDescriptor

descriptor = ReactionCoordinateDescriptor()

train_coords = np.array([0.01, 0.5, 0.8])
train_structures = [get_structure(c) for c in train_coords]
X = np.array(descriptor.get_features(train_structures)).reshape(-1, 1)
Y = np.array([get_true_energy(s) for s in train_structures]).reshape(-1, 1)
```

Let us take a look at the kernel matrix for the training data. Try to interpret the individual entries as similaries of the features. 

Do you see any trends?

```python
K = kernel(X,X, ls=0.1)
print(K)
```

Due to the choice of the magnitude of the kernel we can interpret the kernel entries as percentage similarity between structures. 

Having defined a kernel function we want to train a Gaussian process regression model using it. 
Training a GPR model means solving for the weigth vector $\alpha$ given as
$$\alpha = \left(k(X,X) + \sigma_n^2 I_n\right)^{-1} \left(Y - \mu(X)\right),$$
with $\sigma_n$ a noise hyperparameter that acts a regulariser, and the prior $\mu$ here chosen as the mean function of the $Y$-values. 

Use the `inv` function from `numpy.linalg` to calculated the inverse of `C` below:


```python
# training the model
noise, prior = 0.001, np.mean(Y)
C = K+noise**2*np.eye(K.shape[0])
K_inv = np.linalg.inv(C)
alpha = K_inv @ (Y-prior)

print(alpha)
```

Having solved for `alpha`, we want make predictions using the model. The expression is given as
$$Y_{\text{pred}} = k(X,x^*)^T \alpha,$$
where $x^*$ is the feature vector of the unknown datapoint.

Using `k` and `alpha` calculate Y_pred for `X_test`. Remember to add the prior to the predictions.

```python
# Get our test data
test_coords = np.linspace(0, 1, 100)
test_structures = [get_structure(c) for c in test_coords]
X_test = np.array(descriptor.get_features(test_structures)).reshape(-1, 1)
Y_true = np.array([get_true_energy(s) for s in test_structures]).reshape(-1, 1)

# Calculate kernel similarity between training and test data
k = kernel(X, X_test, ls=0.1)
Y_pred = k.T @ alpha + prior

fig, ax = plt.subplots()
ax.plot(X_test, Y_true, '--g', label='Test data')
ax.plot(X_test, Y_pred, '-b', label='Model prediction')
ax.plot(X, Y, 'xr', label='Training data')
plt.legend(loc='upper right')
display(fig); plt.close(fig)
```

Often you also want the prediction uncertainty for instance in active learning. The expression for the variance of the GPR model is given as
$$Var(Y_{\text{pred}}) = k(x^*, x^*) - k(X,x^*)^T \left(k(X,X) + \sigma_n^2 I_n\right)^{-1} k(X,x^*).$$
Using the variables from previously cells, the standard deviation of the predictions are calculated and plotted.

```python
fig, ax = plt.subplots()
ax.plot(X_test, Y_true, '--g', label='Test data')
ax.plot(X_test, Y_pred, '-b', label='Model prediction')
ax.plot(X, Y, 'xr', label='Training data')

# Calculate uncertainty
std_pred = np.sqrt(np.diag(kernel(X_test)) - np.einsum("ij,ji->i", k.T, K_inv@k)).reshape(-1,1)

ax.fill_between(X_test.flatten(), (Y_pred-std_pred).flatten(), (Y_pred+std_pred).flatten(), alpha=0.1)
plt.legend(loc='upper right')
display(fig); plt.close(fig)
```

Notice how the uncertainty correlates with regions where the GPR model fits the data well and is larger elsewhere.

Now theres some additional structure to the models predictions - though nothing truly different! Thats because while CartesianCoordinates do represent the positions of a structure - it does so in a way that is not particularly useful for learning. 

A good descriptor for learning should make sure that the distance between identical structures is zero. 

```python
# Default
atoms_default = training_data[0].copy()
atoms_default.rattle(0.2)

# Translate a copy of the default
atoms_translated = atoms_default.copy()
atoms_translated.translate([1, 0, 0]) # Translate by 1 Å along x-axis

# Rotate a copy of the default
atoms_rotated = atoms_default.copy()
atoms_rotated.rotate(45, [0, 0, 1]) # Rotate by 45 degrees around z-axis. 

# Permute atoms 0 and 1
atoms_permuted = atoms_default.copy()
P_new = atoms_permuted.positions.copy()
P_new[0, :] = atoms_permuted.positions[1, :]
P_new[1, :] = atoms_permuted.positions[0, :]
atoms_permuted.set_positions(P_new)

test_atoms_list = [atoms_default, atoms_translated, atoms_rotated, atoms_permuted]

fig, axes = plot_structures(test_atoms_list, plot_indices=True)
```

```python
# Calculate features: 
features = np.array(CC_descriptor.get_features(test_atoms_list))

D = np.linalg.norm(features[0]-features, axis=1)

fig, ax = plt.subplots()
x = np.arange(4)
l = ax.bar(x, D, edgecolor='black', alpha=0.5)

ax.set_xticks([0, 1, 2, 3], ['Default', 'Translated', 'Rotated', 'Permuted'])
ax.set_ylabel('Distance to default')
display(fig); plt.close(fig)
```

So the `CartesianCoordinateDescriptor`does not respect the *invariances* that the potential energy surface does. 

#### Exercise: Make a descriptor where each feature is the nearest neighbour distance. 

```python
from bayes_nanospace2025.descriptors import DescriptorBaseClass

class NearestDistanceDescriptor(DescriptorBaseClass):

    descriptor_type = 'global'
    name = 'NearestDistance'

    def __init__(self):
        super().__init__(environment=None)

    def get_number_of_centers(self, atoms):
        return 1

    def create_features(self, atoms):
        distance_matrix = atoms.get_all_distances()
        distance_matrix += np.eye(len(atoms)) * 10
        # Distance matrix is N_atoms x N_atoms. Rows are atoms, columns are distances to other atoms.
        return distance_matrix.min(axis=1)
    


ND_descriptor = NearestDistanceDescriptor()
f = ND_descriptor.get_features(atoms_default)
```

```python
# Calculate features: 
features = np.array(ND_descriptor.get_features(test_atoms_list))

D = np.linalg.norm(features[0]-features, axis=1)

fig, ax = plt.subplots()
x = np.arange(4)
l = ax.bar(x, D, edgecolor='black', alpha=0.5)
ax.set_xticks([0, 1, 2, 3], ['Default', 'Translated', 'Rotated', 'Permuted'])
ax.set_ylabel('Distance to default')
display(fig); plt.close(fig)
```

So this descriptor includes translation and rotational invariance.

Now that we have understood how to make a useful GPR model we can use it in an active learning Bayesian search. 

This invovles choosing an acquisiton function, such that the next training point to update is given by an expression such as 

$$x_{new} =  \mathrm{argmin}_i \ E(x_i)$$

Where $E(x_i)$ is the models prediction about the energy of a structure $x_i$.

We can try such a procedure manually.

```python
from ase.calculators.singlepoint import SinglePointCalculator

# Acquire next structure
structures = [get_structure(v) for v in np.linspace(0, 1, 50)]

model_ouput = [model.predict_energy(s) for s in structures]

i = np.argmin(model_ouput)
x = structures[i]

# Calculate its energy.
E_new = get_true_energy(x)
x.calc = SinglePointCalculator(atoms=x, energy=E_new)

# Add it to the training set
training_energies.append(E_new)
training_data.append(x)

# Train.
model.train(training_data)
```

```python
fig, ax = path_energy_plot(model, training_data)
display(fig); plt.close(fig)
```

The `LowerConfidenceBound`-acquisition function is quite powerful, but there are other options that can be used. 

One such option is the probability of improvement, the figure below illustrates how it works. 

The GPR defines a distribution of functions, the fitting procedure can be thought of as a removing those functions that dont go through (or near enough) the training data. 
When making a prediction the mean value of the functions in the distribution is calculated, whereas the width of the distribution at that point defines the uncertainty. 
This also makes it clear that near training points the uncertainty is small - as all the functions have to yield nearly the same value at the training point. 

In the below figure the blue curve illustrates the prediction of a model, the green point is the best value observed thus far. The orange curve is the distribution around a query point. 
For the probability of improvement the part of the distribution that is below the lowest observed value `f(x')` is integrated, this is illustrated by the shaded area in the below figure. 

```python
from hammer_tutorial_utils.plot import prob_of_imp_figure

fig, ax = prob_of_imp_figure()
display(fig); plt.close(fig)
```

Another acquisition function in the same family is called Expected Improvement, where the expectation value of the improvement rather than just the probability is calculated. 
Note that in both these cases the next point is chosen by maximizing the acquisition function rather than minizing as we have seen so far. 

For both probability of improvement and expected improvement there's a exploitation-exploration controlling hyperparameter $\xi$ that we can think of as replacing $f(x') \rightarrow f(x') - \xi$ in effect pretending that the best observed value of the function is higher than it is. A higher $\xi$ will favor regions with higher uncertainty - at $x'$ the uncertainty is small, thus the distribution is narrow and the probability to find a value better than $f(x') - \xi$ is small - whereas in regions of high uncertainty the distribution is wide with a tail dipping below $f(x') - \xi$. 

Increasing $\xi$ decreases the probability for all query points, but more so for query points with low uncertainty than high uncertainty. 


For a derivation of both of these check this [blog](https://ekamperi.github.io/machine%20learning/2021/06/11/acquisition-functions.html#probability-of-improvement-pi). 

```python
model = make_model(descriptor)

acquisition_functions = []
acquisition_functions.append(LowerConfidenceBound(kappa=2))
acquisition_functions.append(Uncertainty())
acquisition_functions.append(ProbabilityOfImprovement(xi=0.01))
acquisition_functions.append(ExpectedImprovement(xi=0.01))

acquisition_function = acquisition_functions[0]
bayes = BayesianSearch(model=model, acquisition=acquisition_function)

BayesianSearchWidget(bayes, descriptors=descriptors, acquisition_functions=acquisition_functions)
```
